```python
import nibabel as ni

import sklearn.linear_model as lm
import osmosis as oz
import osmosis.model.analysis as oza
import osmosis.model.fiber as mtm
import osmosis.model.dti as dti
import osmosis.viz.mpl as mpl
import osmosis.utils as ozu

import osmosis.io as oio
oio.data_path = '/biac4/wandell/biac2/wandell6/data/arokem/osmosis'

data_path = oz.__path__[0] + '/data/'

dwi1 = '0005_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.nii.gz'
dwi2 = '0007_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.nii.gz'

bvecs1 = '0005_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.bvecs'
bvecs2 = '0007_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.bvecs'

bvals1 = '0005_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.bvals'
bvals2 = '0007_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.bvals'


TM1 = dti.TensorModel(data_path + dwi1, 
                      data_path + bvecs1, 
                      data_path + bvals1)


TM2 = dti.TensorModel(data_path + dwi2, 
                      data_path + bvecs2, 
                      data_path + bvals2)

t1_ni = ni.load(data_path + 'FP_t1.nii.gz')
t1_data = t1_ni.get_data()
t1_aff = np.matrix(t1_ni.get_affine())

t1_data.shape
```
