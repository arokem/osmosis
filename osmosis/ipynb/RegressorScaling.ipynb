```python
import osmosis.utils as ozu
```

```python
reg1 = np.sin(np.linspace(0,np.pi*6,100)) + 2 + np.random.randn(100) * 0.1reg1 = reg1/np.max(reg1)reg2 = np.cos(np.linspace(0,np.pi*6,100)) + 2 + np.random.randn(100) * 0.1reg2 = reg2/np.max(reg2)constant = (np.random.rand() + np.zeros(100))/2 + 0.5
```

```python
plot(reg1)plot(reg2)
```

```python
y = 0.5 * reg1 + 0.7 * reg2 + constant
```

```python
plot(y)
```

```python
A = np.vstack([reg1, reg2, np.ones(100)/100])
```

```python
beta = np.asarray(np.dot(ozu.ols_matrix(A).T, y)).squeeze()
```

```python
y_hat = beta[0] * reg1 + beta[1] * reg2 + beta[2] *np.ones(100)
```

```python
plot(y)plot(y_hat)
```
