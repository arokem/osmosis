```python
n_sims = 100
bval = 1000
subject = 'FP'
data1, data2 =  oio.get_dwi_data(bval, subject)
mask = oio.get_wm_mask('/Users/arokem/projects/data_osmosis_raw/FP/FP_wm_mask.nii.gz', resample=data1[0])
```

```python
DWI1 = dwi.DWI(*data1, mask=mask)
DWI2 = dwi.DWI(*data2, mask=mask)
noise_sig = (DWI1.data[DWI1.mask]- DWI2.data[DWI2.mask])/2.0
fig = viz.probability_hist(ozu.rms(noise_sig))
S0 = np.mean([np.median(DWI1._flat_S0), np.median(DWI2._flat_S0)])
```

```python
n_bvecs = 150
bvecs = np.loadtxt(data1[1])
bvals = np.loadtxt(data1[2])
b_idx = np.where(bvals>0)[0]
```

```python
vec1 = bvecs[:,b_idx[0]]
```

```python
angles = np.linspace(10, 90, 11)
weights = np.linspace(0.5, 1.0, 9)
sphere = np.linspace(S0,S0,1)
vol = np.empty((angles.shape[0], weights.shape[0], sphere.shape[0], n_bvecs + 10))
```

```python
for i, angle in enumerate(angles):
    # Rotate around the null-space of the vector and its inverse: 
    R = geo.rodriguez_axis_rotation(ozu.null_space([vec1, -vec1, [0,0,0]])[:,0], angle)
    # We should generate a signal that can be reproduced *exactly*, so we find a vector that is closest
    # To the vector that has the angle that we want it to have. Start with a candidate:
    vec2= np.array(np.dot(R.T, vec1)).squeeze()
    angles[i] = np.rad2deg(ozu.vector_angle(vec1, vec2))
    odf_vecs = np.array([vec1, vec2])
    for j, w in enumerate(weights):
        for k, sph in enumerate(sphere):
            odf = sim.ODF(odf_vecs, [w, 1-w])
            v = sim.Voxel(bvecs[:,b_idx], bvals[:, b_idx], odf)
            S0 = 0 + np.ones(10) * sph
            vol[i,j,k] = np.concatenate((S0, v.signal(np.mean(S0)))) 
```

```python
cluster_angs1 = []
cluster_angs2 = []

for i in xrange(n_sims): 
    
    noise_idx1 = int(np.random.rand() * noise_sig.shape[0])
    vol_noise1 = np.abs(vol + noise_sig[noise_idx1])
    
    noise_idx2 = int(np.random.rand() * noise_sig.shape[0])
    # Just making sure the two noise samples are actually different:
    while noise_idx1 == noise_idx2:
        noise_idx2 = int(np.random.rand() * sig_diff.shape[0])

    vol_noise2 = np.abs(vol + noise_sig[noise_idx2])
        
    
    TM1 = dti.TensorModel(vol_noise1,
                         bvecs,
                         bvals,
                         params_file = 'temp')

    TM2 = dti.TensorModel(vol_noise2,
                         bvecs,
                         bvals,
                         params_file = 'temp')

    SFM1 = ssd.SparseDeconvolutionModel(vol_noise1,
                                        bvecs,
                                        bvals,
                                        axial_diffusivity=1.5,
                                        radial_diffusivity=0.5,
                                        params_file = 'temp')

    SFM2 = ssd.SparseDeconvolutionModel(vol_noise2,
                                        bvecs,
                                        bvals,
                                        axial_diffusivity=1.5,
                                        radial_diffusivity=0.5,
                                        params_file = 'temp')

    cluster_angs1.append(np.rad2deg(ozu.vector_angle(*SFM1.cluster_fodf[-1,0,0][:2])))
    cluster_angs2.append(np.rad2deg(ozu.vector_angle(*SFM2.cluster_fodf[-1,0,0][:2])))
```
