```python
data_path = os.path.split(oz.__file__)[0] + '/data/'
T1 = ni.load(data_path + 'FP_t1_resampled_to_dwi.nii.gz').get_data()
T1[np.isnan(T1)]=0
wm = ni.load(data_path + 'FP_white_matter_resamp_to_dwi.nii.gz').get_data()
wm_mask = np.zeros(wm.shape)
wm_mask[np.where(wm==1)] = 1

dwi_4k_1 = '0005_01_DWI_2mm150dir_2x_b4000_aligned_trilin.nii.gz'
dwi_4k_2 = '0007_01_DWI_2mm150dir_2x_b4000_aligned_trilin.nii.gz'

bvecs_4k_1 = '0005_01_DWI_2mm150dir_2x_b4000_aligned_trilin.bvecs'
bvecs_4k_2 = '0007_01_DWI_2mm150dir_2x_b4000_aligned_trilin.bvecs'

bvals_4k_1 = '0005_01_DWI_2mm150dir_2x_b4000_aligned_trilin.bvals'
bvals_4k_2 = '0007_01_DWI_2mm150dir_2x_b4000_aligned_trilin.bvals'


TM_4k_1 = ozm.TensorModel(data_path + dwi_4k_1, 
                          data_path + bvecs_4k_1, 
                          data_path + bvals_4k_1,
                          mask=wm_mask)

TM_4k_2 = ozm.TensorModel(data_path + dwi_4k_2, 
                          data_path + bvecs_4k_2, 
                          data_path + bvals_4k_2,
                          mask=wm_mask)


dwi_2k_1 = '0005_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.nii.gz'
dwi_2k_2 = '0007_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.nii.gz'

bvecs_2k_1 = '0005_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.bvecs'
bvecs_2k_2 = '0007_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.bvecs'

bvals_2k_1 = '0005_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.bvals'
bvals_2k_2 = '0007_01_DTI_2mm_150dir_2x_b2000_aligned_trilin.bvals'


TM_2k_1 = ozm.TensorModel(data_path + dwi_2k_1, 
                          data_path + bvecs_2k_1, 
                          data_path + bvals_2k_1, 
                          mask=wm_mask)


TM_2k_2 = ozm.TensorModel(data_path + dwi_2k_2, 
                          data_path + bvecs_2k_2, 
                          data_path + bvals_2k_2,
                          mask=wm_mask)

dwi_1k_1 = '0009_01_DWI_2mm150dir_2x_b1000_aligned_trilin.nii.gz'
dwi_1k_2 = '0011_01_DWI_2mm150dir_2x_b1000_aligned_trilin.nii.gz'

bvecs_1k_1 = '0009_01_DWI_2mm150dir_2x_b1000_aligned_trilin.bvecs'
bvecs_1k_2 = '0011_01_DWI_2mm150dir_2x_b1000_aligned_trilin.bvecs'

bvals_1k_1 = '0009_01_DWI_2mm150dir_2x_b1000_aligned_trilin.bvals'
bvals_1k_2 = '0011_01_DWI_2mm150dir_2x_b1000_aligned_trilin.bvals'


TM_1k_1 = ozm.TensorModel(data_path + dwi_1k_1, 
                          data_path + bvecs_1k_1, 
                          data_path + bvals_1k_1, 
                          mask=wm_mask)


TM_1k_2 = ozm.TensorModel(data_path + dwi_1k_2, 
                          data_path + bvecs_1k_2, 
                          data_path + bvals_1k_2,
                          mask=wm_mask)
```

```python
b0_1k = np.concatenate([TM_1k_1.data[..., TM_1k_1.b0_idx], TM_1k_2.data[..., TM_1k_2.b0_idx]], -1) 
b0_2k = np.concatenate([TM_2k_1.data[..., TM_2k_1.b0_idx], TM_1k_2.data[..., TM_2k_2.b0_idx]], -1)
b0_4k = np.concatenate([TM_4k_1.data[..., TM_4k_1.b0_idx], TM_4k_2.data[..., TM_4k_2.b0_idx]], -1) 

snr_0 = []
snr_1k = [] 
snr_2k = []
snr_4k = []
for b0 in [b0_1k, b0_2k, b0_4k]:
    sig = np.mean(b0, -1)
    noise = np.std(b0, -1)
    snr_0.append(sig/noise)
    snr_1k.append(sig/noise * np.exp(-1 * TM_1k_1.mean_diffusivity))
    snr_2k.append(sig/noise * np.exp(-2 * TM_1k_1.mean_diffusivity))
    snr_4k.append(sig/noise * np.exp(-4 * TM_1k_1.mean_diffusivity))

snr_0 = np.mean(snr_0,0)
snr_1k = np.mean(snr_1k,0)
snr_2k = np.mean(snr_2k,0)
snr_4k = np.mean(snr_4k,0)
```

```python
fig_snr_hist, ax = plt.subplots(1)
lower = []
median = []
upper = []

for label, snr in zip(['0', '1000', '2000', '4000'], [snr_0, snr_1k, snr_2k, snr_4k]):
    vol = ozu.nans(snr.shape)
    fig = viz.mosaic(T1.T[29:60][::2], cmap=matplotlib.cm.bone, cbar=False)
    vol[np.isfinite(snr)] = snr[np.isfinite(snr)]
    fig = viz.mosaic(vol.T[29:60][::2], cmap=matplotlib.cm.hot, fig=fig)
    fig.set_size_inches([15,12])
    fig_snr_hist = viz.probability_hist(vol[np.isfinite(vol)], fig=fig_snr_hist, label=label)
    l,m,u = [stats.scoreatpercentile(vol[np.isfinite(vol)], s) for s in 2.5, 50, 97.5]
    lower.append(l)
    median.append(m)
    upper.append(u)


fig_medians, ax_medians = plt.subplots(1)
ax_medians.errorbar([0,1,2,3], median, yerr=[lower,upper])
ax_medians.figure.set_size_inches([12,8])
ax_medians.set_xlim([-0.5, 3.5])
ax_medians.set_xticks([0,1,2,3])
ax_medians.set_xticklabels([0,1000,2000,4000])
ax_medians.semilogy()
ax_medians.grid()
ax_medians.set_xlabel('b')
ax_medians.set_ylabel('SNR')
fig_medians.savefig('/Users/arokem/SNR.svg')
```

```python
corr1k_flat=np.empty(TM_1k_1._flat_signal.shape[0])
corr2k_flat=np.empty(TM_2k_1._flat_signal.shape[0])
corr4k_flat=np.empty(TM_4k_1._flat_signal.shape[0])
diff1k_flat=np.empty(TM_1k_1._flat_signal.shape)
diff2k_flat=np.empty(TM_2k_1._flat_signal.shape)
diff4k_flat=np.empty(TM_4k_1._flat_signal.shape)
mean1k_flat=np.empty(TM_1k_1._flat_signal.shape)
mean2k_flat=np.empty(TM_2k_1._flat_signal.shape)
mean4k_flat=np.empty(TM_4k_1._flat_signal.shape)


for vox in range(corr1k_flat.shape[0]):
   corr1k_flat[vox] = np.corrcoef(TM_1k_1._flat_signal[vox], TM_1k_2._flat_signal[vox])[0,1] ** 2
   corr2k_flat[vox] = np.corrcoef(TM_2k_1._flat_signal[vox], TM_2k_2._flat_signal[vox])[0,1] ** 2
   corr4k_flat[vox] = np.corrcoef(TM_4k_1._flat_signal[vox], TM_4k_2._flat_signal[vox])[0,1] ** 2
   diff1k_flat[vox] = np.abs(TM_1k_1._flat_signal[vox] - TM_1k_2._flat_signal[vox])
   diff2k_flat[vox] = np.abs(TM_2k_1._flat_signal[vox] - TM_2k_2._flat_signal[vox])
   diff4k_flat[vox] = np.abs(TM_4k_1._flat_signal[vox] - TM_4k_2._flat_signal[vox])
   mean1k_flat[vox] = (TM_1k_1._flat_signal[vox] + TM_1k_2._flat_signal[vox])/2.
   mean2k_flat[vox] = (TM_2k_1._flat_signal[vox] + TM_2k_2._flat_signal[vox])/2.
   mean4k_flat[vox] = (TM_4k_1._flat_signal[vox] + TM_4k_2._flat_signal[vox])/2.
    
corr1k = ozu.nans(TM_1k_1.signal.shape[:3])
corr1k[TM_1k_1.mask] = corr1k_flat
    
corr2k = ozu.nans(TM_2k_1.signal.shape[:3])
corr2k[TM_2k_1.mask] = corr2k_flat

corr4k = ozu.nans(TM_4k_1.signal.shape[:3])
corr4k[TM_4k_1.mask] = corr4k_flat

```
