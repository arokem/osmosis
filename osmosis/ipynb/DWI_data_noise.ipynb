```python
all_b0 = np.concatenate([TM_1k_1.data[..., TM_1k_1.b0_idx], TM_1k_2.data[..., TM_1k_2.b0_idx], 
                         TM_2k_1.data[..., TM_2k_1.b0_idx], TM_1k_2.data[..., TM_2k_2.b0_idx],
                         TM_4k_1.data[..., TM_4k_1.b0_idx], TM_4k_2.data[..., TM_4k_2.b0_idx]], -1) 

sig = np.mean(all_b0, -1)
noise = np.std(all_b0, -1)
snr_0 = sig/noise
snr_0[wm_mask==0] = np.nan
snr_1000 = sig/noise * np.exp(-1 * TM_1k_1.mean_diffusivity)
snr_2000 = sig/noise * np.exp(-2 * TM_1k_1.mean_diffusivity)
snr_4000 = sig/noise * np.exp(-4 * TM_1k_1.mean_diffusivity)
```

```python
corr1k_flat=np.empty(TM_1k_1._flat_signal.shape[0])
corr2k_flat=np.empty(TM_2k_1._flat_signal.shape[0])
corr4k_flat=np.empty(TM_4k_1._flat_signal.shape[0])
diff1k_flat=np.empty(TM_1k_1._flat_signal.shape)
diff2k_flat=np.empty(TM_2k_1._flat_signal.shape)
diff4k_flat=np.empty(TM_4k_1._flat_signal.shape)
mean1k_flat=np.empty(TM_1k_1._flat_signal.shape)
mean2k_flat=np.empty(TM_2k_1._flat_signal.shape)
mean4k_flat=np.empty(TM_4k_1._flat_signal.shape)


for vox in range(corr1k_flat.shape[0]):
   corr1k_flat[vox] = np.corrcoef(TM_1k_1._flat_signal[vox], TM_1k_2._flat_signal[vox])[0,1] ** 2
   corr2k_flat[vox] = np.corrcoef(TM_2k_1._flat_signal[vox], TM_2k_2._flat_signal[vox])[0,1] ** 2
   corr4k_flat[vox] = np.corrcoef(TM_4k_1._flat_signal[vox], TM_4k_2._flat_signal[vox])[0,1] ** 2
   diff1k_flat[vox] = np.abs(TM_1k_1._flat_signal[vox] - TM_1k_2._flat_signal[vox])
   diff2k_flat[vox] = np.abs(TM_2k_1._flat_signal[vox] - TM_2k_2._flat_signal[vox])
   diff4k_flat[vox] = np.abs(TM_4k_1._flat_signal[vox] - TM_4k_2._flat_signal[vox])
   mean1k_flat[vox] = (TM_1k_1._flat_signal[vox] + TM_1k_2._flat_signal[vox])/2.
   mean2k_flat[vox] = (TM_2k_1._flat_signal[vox] + TM_2k_2._flat_signal[vox])/2.
   mean4k_flat[vox] = (TM_4k_1._flat_signal[vox] + TM_4k_2._flat_signal[vox])/2.
    
corr1k = ozu.nans(TM_1k_1.signal.shape[:3])
corr1k[TM_1k_1.mask] = corr1k_flat
    
corr2k = ozu.nans(TM_2k_1.signal.shape[:3])
corr2k[TM_2k_1.mask] = corr2k_flat

corr4k = ozu.nans(TM_4k_1.signal.shape[:3])
corr4k[TM_4k_1.mask] = corr4k_flat

```
