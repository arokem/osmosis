```python
import nibabel as nib
import dipy.reconst.dti as dti
import dipy.core.gradients as grad
import dipy.data as dpd
import dipy.sims.voxel as sims
import dipy.core.sphere as sph
import osmosis.utils as ozu
```

```python
sphere = dpd.get_sphere('symmetric362')
bvecs = np.concatenate(([[0, 0, 0]], sphere.vertices))
bvals = np.zeros(len(bvecs)) + 1000
bvals[0] = 0
gtab = grad.gradient_table(bvals, bvecs)
mevals = np.array(([0.0015, 0.0015, 0.0001], [0.0015, 0.0003, 0.0001]))
mevecs = [ np.array( [ [1, 0, 0], [0, 1, 0], [0, 0, 1] ] ),
           #np.array( [ [1, 0, 0], [0, 1, 0], [0, 0, 1] ] )
           np.array( [ [0, 0, 1], [1, 0, 0], [0, 1, 0] ] )
           ]

S1 = sims.single_tensor( gtab, 100, mevals[0], mevecs[0], snr=None )
S2 = sims.single_tensor( gtab, 100, mevals[1], mevecs[1], snr=None )

dtm = dti.TensorModel(gtab, 'LS')
dtf1 = dtm.fit(S1)
dtf2 = dtm.fit(S2)

odf1 = dtf1.odf(sphere)
odf2 = dtf2.odf(sphere)
```

```python
vertices = gtab.bvecs[~gtab.b0s_mask]
```

```python
sphere = sph.Sphere(xyz=vertices)
```

```python
ang = np.rad2deg(np.arccos(np.dot(sphere.vertices, sphere.vertices.T)))
ang[np.isnan(ang)] = 0
ang_rav = ang.ravel()
```

```python
plot(odf1)
plot(odf2)
```

```python
dotS1S2 = np.dot(S1.reshape(-1,1), S2.reshape(1,-1))
dotS1 = S1.reshape(-1, 1) + np.ones((S1.shape[0], S1.shape[0]))
dotS2 = S2.reshape(1, -1) + np.ones((S2.shape[0], S2.shape[0]))
corr = dotS1S2/(dotS1 * dotS2)
```

```python
fig, ax = plt.subplots(1)
im = ax.matshow(dotS1S2)
fig.colorbar(im)

fig, ax = plt.subplots(1)
im = ax.matshow(dotS1)
fig.colorbar(im)

fig, ax = plt.subplots(1)
im = ax.matshow(dotS2)
fig.colorbar(im)

fig, ax = plt.subplots(1)
im = ax.matshow(corr)
fig.colorbar(im)
```

```python
n = 180
step = 180./n
degs = np.linspace(0, 180, n)
cc = np.zeros(len(degs))
idx = []
for ii, d in enumerate(degs):
    idx.append(np.where(np.logical_and(ang_rav>=d, ang_rav<=d+step)))
    cc[ii] = np.mean(corr_rav[idx[-1]])
```

```python
plot(degs, cc, 'o')
```
