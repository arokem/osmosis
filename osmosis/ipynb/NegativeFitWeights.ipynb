For a while now, I have been wondering what negative weights on tensors/fibers might entail. Franco brought this up again today. Jason mentioned an assertion I had previously made that a negative weight is a fiber running orthogonal to the original fiber ("Van Wedeen would predict that you would have negative weights everywhere!"), but I was never to sure about this assertion and decided to test this with a simulated signal. 


Here goes. We start by importing the name-spaces we need for this: 

We use dipy for the tensor fitting code and for some of their simulation code (Rician noise generator):

```python
import dipy.reconst.dti as dti
import dipy.sims.phantom as dps
```

We use my own stuff for vizualization and simulation. "utils" are various useful things: 

```python
import osmosis.viz as viz
import osmosis.simulation as sim
import osmosis.utils as ozu
```

We get b vectors from the camino electro-static repulsion point set and set the bvalues to have the same length and be equal to 2000: 

```python
bvecs = ozu.get_camino_pts(150)
bvals = np.ones(bvecs.shape[-1]) * 2000
```

The following creates a simulated ODF, with 1 population of fibers pointing in the direction of $(1,0,0)$

```python
odf = sim.ODF([[1,0,0]],[1])
```

The following creates a simulated voxel measurment with that odf: 

```python
v = sim.Voxel(bvecs, bvals, odf)
```

We generate the signal, assuming that $S_0 = 100$  and display it on a projection: 

```python
sig = v.signal(100)
sig = dps.add_noise(sig, noise_type='rician', snr=5) 
fig = viz.sig_on_projection(bvecs, sig, tri=True)
```

To generate the inverse of this signal, I demean the original signal, multiply by $-1$ and then add back the mean: 

```python
new_sig = (-1 * (sig - np.mean(sig))) + np.mean(sig)
```

This is what that looks like: 

```python
fig = viz.sig_on_projection(bvecs, new_sig, tri=True)
```

Hmm. Looks like a pancake... 

Let's go and fit tensors to that: 

```python
T1 = dti.Tensor(np.hstack([500, sig]), np.hstack([0, bvals]), np.vstack([[0,0,0], bvecs.T]))
T2 = dti.Tensor(np.hstack([500, new_sig]), np.hstack([0, bvals]), np.vstack([[0,0,0], bvecs.T]))
```

We quantify the transformation using the Westin statistics 

```python
print ozu.tensor_linearity(*T1.evals)
print ozu.tensor_planarity(*T1.evals)
```

```python
print ozu.tensor_linearity(*T2.evals)
print ozu.tensor_planarity(*T2.evals)
```

Yep - while our first tensor is rather linear, the second tensor is more planar.

Conclusion: a single tensor multiplied by a is a pancake, not an orthogonal tensor.
