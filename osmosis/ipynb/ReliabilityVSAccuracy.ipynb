```python
import osmosis.utils as ozu
```

```python
n = 1000
true_params = -1,10,0
```

```python
x = np.random.rand(n) * 10
```

```python
y = np.polyval(true_params, x) + np.random.randn(n) * 4
rmse_data = ozu.rmse(y, np.polyval(true_params, x))
```

```python
boot_n = 100
boot_params_linear = []
cod_linear = []
rmse_model_linear = []

boot_params_quad = []
cod_quad = []
rmse_model_quad = []

for i in range(boot_n):
    idx_boot1 = np.random.random_integers(0,n-1,n)
    idx_boot2 = np.random.random_integers(0,n-1,n)
    
    params_linear = np.polyfit(x[idx_boot1], y[idx_boot1], 1)
    boot_params_linear.append(params_linear)
    
    y_hat_linear = np.polyval(params_linear, x[idx_boot2])
    cod_linear.append(ozu.coeff_of_determination(y[idx_boot2],y_hat_linear))
    rmse_model_linear.append(ozu.rmse(y[idx_boot2],y_hat_linear))
    
    params_quad = np.polyfit(x[idx_boot1], y[idx_boot1], 2)
    boot_params_quad.append(params_quad)
    
    y_hat_quad = np.polyval(params_quad, x[idx_boot2]) 
    cod_quad.append(ozu.coeff_of_determination(y[idx_boot2], y_hat_quad))
    rmse_model_quad.append(ozu.rmse(y[idx_boot2], y_hat_quad))
    
    
    
```

```python
fig, ax = plt.subplots(1)
ax.plot(x,y,'o', label='data', color='w')
ax.plot(np.linspace(0,10,100), np.polyval(true_params, np.linspace(0,10,100)),'k--', linewidth=2)

all_linear = []
for params in boot_params_linear:
    this_linear =  np.polyval(params, np.linspace(0,10,100))
    all_linear.append(this_linear)

# Sort in each value of x:
sort_linear = np.sort(all_linear, 0) 

ax.fill_between(np.linspace(0,10,100), sort_linear[4],sort_linear[-5], alpha=0.2, color='k')
ax.plot(np.linspace(0,10,100), sort_linear[49], color='k')


ax.set_xlabel('x')
ax.set_ylabel('y')

fig.savefig('/Users/arokem/projects/osmosis/doc/figures/Figure1_a.svg')

fig, ax = plt.subplots(1)
ax.plot(x,y,'o')
ax.plot(np.linspace(0,10,100), np.polyval(true_params, np.linspace(0,10,100)),'k--', linewidth=2)

for params in boot_params_quad:
    
    ax.plot(np.linspace(0,10,100), np.polyval(params, np.linspace(0,10,100)))

ax.set_xlabel('x')
ax.set_ylabel('y')
fig.savefig('/Users/arokem/projects/osmosis/doc/figures/Figure1_a.svg')
```
