```python
target = as_image(data_path + 'jw_t1_class.nii.gz')
source = as_image(data_path + 'R_squared.nii.gz')

# This it the transformation between source and target:
source2target = np.matrix([[0.8915, -0.0226, 0.2890, 1.6637],
                     [0.0731, -0.0035, -3.5330, 276.5624],
                     [-0.0235, -0.8938, 0.0064, 243.9781],
                     [0, 0, 0, 1.0000]])
mapping = np.dot(np.dot(target.affine, source2target), source.affine)

resimg = np.asarray(resample.resample(source, target.coordmap, mapping, target.shape, order=0))
```
