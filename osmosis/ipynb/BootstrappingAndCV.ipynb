```python
import osmosis.utils as ozu
```

```python
n = 1000true_params = -1,10,0
```

```python
x = np.random.rand(n) * 10
```

```python
y = np.polyval(true_params, x) + np.random.randn(n) * 4rmse_data = ozu.rmse(y, np.polyval(true_params, x))
```

```python
boot_n = 100boot_params_linear = []cod_linear = []rmse_model_linear = []boot_params_quad = []cod_quad = []rmse_model_quad = []for i in range(boot_n):    idx_boot1 = np.random.random_integers(0,n-1,n)    idx_boot2 = np.random.random_integers(0,n-1,n)        params_linear = np.polyfit(x[idx_boot1], y[idx_boot1], 1)    boot_params_linear.append(params_linear)        y_hat_linear = np.polyval(params_linear, x[idx_boot2])    cod_linear.append(ozu.coeff_of_determination(y[idx_boot2],y_hat_linear))    rmse_model_linear.append(ozu.rmse(y[idx_boot2],y_hat_linear))        params_quad = np.polyfit(x[idx_boot1], y[idx_boot1], 2)    boot_params_quad.append(params_quad)        y_hat_quad = np.polyval(params_quad, x[idx_boot2])     cod_quad.append(ozu.coeff_of_determination(y[idx_boot2], y_hat_quad))    rmse_model_quad.append(ozu.rmse(y[idx_boot2], y_hat_quad))            
```
