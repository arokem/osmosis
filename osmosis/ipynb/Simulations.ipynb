```python
bvecs = ozu.get_camino_pts(150)
bvals = np.ones(bvecs.shape[-1]) * 2000
```

```python
r, phi, theta= geo.cart2sphere(bvecs[0,0], bvecs[1,0], bvecs[2,0])
```

```python
angles = np.linspace(10, 90, 5)
weights = np.linspace(0, 0.5, 5)
sphere = np.linspace(100, 500, 5)
vol = np.empty((angles.shape[0], weights.shape[0], sphere.shape[0], 160))
```

```python
for i, angle in enumerate(angles):
    x,y,z = geo.sphere2cart(r, theta + deg2rad(angle), phi) 
    vec = [x,y,z]
    odf_vecs = np.array([bvecs[:,0], vec])
    for j, w in enumerate(weights):
        for k, sph in enumerate(sphere):
            odf = sim.ODF(odf_vecs, [w, 1-w])
            v = sim.Voxel(bvecs, bvals, odf)
            S0 = 100 + np.ones(10) * sph
            vol[i,j,k] = np.concatenate((S0, v.signal(np.mean(S0)))) 

vol1 = vol
#dps.add_rician_noise(vol, 1000)
vol2 = dps.add_rician_noise(vol, 1000)

vol.shape, weights, angles
```

```python
fac = 4.
cod = np.zeros(np.ceil(90/fac))

for j in range(CTM.predict_all.shape[-2]):
    # viz.sig_on_projection(bvecs, CTM.predict_all[i][j])
    ang = np.rad2deg(ozu.vector_angle(CTM.bvecs[:,CTM.b_idx].T[CTM.rot_idx[j][0]], 
                                      CTM.bvecs[:,CTM.b_idx].T[CTM.rot_idx[j][1]]))
    ang = np.min([ang, 180-ang])
    this_cod = ozu.coeff_of_determination(CTM.predict_all[vox_idx][j], CTM.signal[vox_idx])
    if np.isfinite(this_cod):
        if this_cod>cod[np.floor(ang/fac)]:
            cod[np.floor(ang/fac)] = this_cod

fig,ax = plt.subplots(1)
ax.plot(cod)
ax.set_xlabel('Angle')
t = ax.get_xticks()
ax.set_xticklabels([str(x*fac) for x in t])

print CTM.fit_angle[vox_idx]
```
