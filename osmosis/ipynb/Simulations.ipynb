```python
bvecs = ozu.get_camino_pts(150)
bvals = np.ones(bvecs.shape[-1]) * 2000
```

```python
vec1 = bvecs[:,1]
```

```python
angles = np.linspace(10, 90, 5)
weights = np.linspace(0, 0.5, 5)
sphere = np.linspace(100, 500, 5)
vol = np.empty((angles.shape[0], weights.shape[0], sphere.shape[0], 160))
```

```python
for i, angle in enumerate(angles):
    #print angle
    R = geo.rodriguez_axis_rotation([0,1,0], angle)
    vec2 = np.array(np.dot(R.T, vec1)).squeeze()
    odf_vecs = np.array([vec1, vec2])
    for j, w in enumerate(weights):
        for k, sph in enumerate(sphere):
            odf = sim.ODF(odf_vecs, [w, 1-w])
            v = sim.Voxel(bvecs, bvals, odf)
            S0 = 100 + np.ones(10) * sph
            vol[i,j,k] = np.concatenate((S0, v.signal(np.mean(S0)))) 

vol1 = dps.add_noise(vol, snr=10, noise_type='rician')
vol2 = dps.add_noise(vol, snr=10, noise_type='rician')


#angles, weights, sphere
vol1.shape
```
