# Modeling diffusion in 1D

This is based on equations in Dell'acqua et al.(2010). 

Under the assumption of slow exchange between fiber compartments in the voxel, The normalized signal $S(\theta, b) = \frac{S_b(\theta)}{S_0}$ is modeled as: 
    
 $S(\theta,b) = \sum_{i=1}^{m}{f_i e^{-b (\lambda_i cos^2(\theta-\phi_i) + \beta_i sin^2(\theta-\phi_i)}} + \sum_{j=1}^{n}{f_j e ^{-b\beta_j}}$
                                        
Where the left side of the equation represents the fiber contributions and the right side represents contributions from isotropic components. Each $\phi_i$ is the direction of a distinct fiber population, $\lambda_i$ and $\beta_i$ represent the axial and radial diffusivity of each fiber response function and $f_i$ represents the weight on that fiber population. 

This can be re-written more compactly as: 

 $S(\theta,b) = \sum_{i=1}^{m}{c_i e^{-b\alpha_i cos(\theta - \phi_i)}} + \sum_{j=1}^{q}{c_j}$
 
Where $\alpha_i = \lambda_i - \beta_i$ and $c_i = f_i e^{-b\beta_i}$
