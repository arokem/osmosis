```python
import os

import nibabel as ni
import osmosis.viz as viz
import osmosis as oz
import osmosis.utils as ozu
```

```python
data_path = os.path.split(oz.__file__)[0] + '/data/'

Models = ['TensorModel', 'CanonicalTensorModel']

PDD = {}

for Model in Models:
    PDD[Model] = {}
    PDD[Model]['pdd4k_1'] = '0007_01_DWI_2mm150dir_2x_b4000_aligned_trilin%sPDD.nii.gz'%Model
    PDD[Model]['pdd4k_2'] = '0005_01_DWI_2mm150dir_2x_b4000_aligned_trilin%sPDD.nii.gz'%Model
    PDD[Model]['pdd2k_1'] = '0007_01_DTI_2mm_150dir_2x_b2000_aligned_trilin%sPDD.nii.gz'%Model
    PDD[Model]['pdd2k_2'] = '0005_01_DTI_2mm_150dir_2x_b2000_aligned_trilin%sPDD.nii.gz'%Model
    PDD[Model]['pdd1k_1'] = '0011_01_DWI_2mm150dir_2x_b1000_aligned_trilin%sPDD.nii.gz'%Model
    PDD[Model]['pdd1k_2'] = '0009_01_DWI_2mm150dir_2x_b1000_aligned_trilin%sPDD.nii.gz'%Model
```

```python
wm_mask = ni.load(data_path + 'FP_white_matter_resamp_to_dwi.nii.gz').get_data()
wm_idx = np.where(wm_mask)
```

```python
for Model in Models:
    pdd1k_1_d = np.abs(ni.load(data_path + PDD[Model]['pdd1k_1']).get_data())[wm_idx]
    pdd1k_2_d = np.abs(ni.load(data_path + PDD[Model]['pdd1k_2']).get_data())[wm_idx]

    pdd2k_1_d = np.abs(ni.load(data_path + PDD[Model]['pdd2k_1']).get_data())[wm_idx]
    pdd2k_2_d = np.abs(ni.load(data_path + PDD[Model]['pdd2k_2']).get_data())[wm_idx]

    pdd4k_1_d = np.abs(ni.load(data_path + PDD[Model]['pdd2k_1']).get_data())[wm_idx]
    pdd4k_2_d = np.abs(ni.load(data_path + PDD[Model]['pdd2k_2']).get_data())[wm_idx]
    
    to_hist1k_1k = []
    to_hist1k_2k = []
    
    for vox in range(len(pdd1k_1_d)):
        to_hist1k_1k.append(ozu.vector_angle(pdd1k_1_d[vox], pdd1k_2_d[vox]))
        to_hist1k_2k.append(ozu.vector_angle(pdd1k_1_d[vox], pdd2k_1_d[vox]))
    
    to_hist1k_1k = np.array(to_hist1k_1k)
    fig = viz.probability_hist(np.rad2deg(to_hist1k_1k[np.isfinite(to_hist1k_1k)]))

    to_hist1k_2k = np.array(to_hist1k_2k)
    fig = viz.probability_hist(np.rad2deg(to_hist1k_2k[np.isfinite(to_hist1k_2k)]))
```
