```python
nSubjects = 25
nRegions = 20; 
d = np.random.randn(nRegions, nSubjects)
```

```python
u,s,v = np.linalg.svd(d, full_matrices=True)
```

```python
v.shape
```

```python
# u and v are orthonormal matrices: 
matshow(np.dot(u,u.T))
matshow(np.dot(v,v.T))

# s can be thought of as a diagonal matrix: 
matshow(np.diag(s))

# The entries of S are the singular values - these are closely related to the variance accounted for by the principal components
fig, ax = plt.subplots()
ax.plot(s)
```
