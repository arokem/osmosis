```python
data_path = os.path.split(oz.__file__)[0] + '/data/'

MCT1k = ni.load(data_path + 'PointyMultiCanonicalTensorModel_relative_rmse_b1000.nii.gz').get_data()
MCT2k = ni.load(data_path + 'PointyMultiCanonicalTensorModel_relative_rmse_b2000.nii.gz').get_data()
MCT4k = ni.load(data_path + 'PointyMultiCanonicalTensorModel_relative_rmse_b4000.nii.gz').get_data()

CT1k = ni.load(data_path + 'PointyCanonicalTensorModel_relative_rmse_b1000.nii.gz').get_data()
CT2k = ni.load(data_path + 'PointyCanonicalTensorModel_relative_rmse_b2000.nii.gz').get_data()
CT4k = ni.load(data_path + 'PointyCanonicalTensorModel_relative_rmse_b4000.nii.gz').get_data()

wm_idx = np.where(ni.load(data_path + 'FP_white_matter_resamp_to_dwi.nii.gz').get_data()==1)
t1 = ni.load(data_path + 'FP_t1_resampled_to_dwi.nii.gz').get_data()
t1[~np.isfinite(t1)]=0
```

```python
diff_1k = MCT1k - CT1k
diff_2k = MCT2k - CT2k
diff_4k = MCT4k - CT4k

diff_1k_wm = ozu.nans(diff_1k.shape)
diff_2k_wm = ozu.nans(diff_2k.shape)
diff_4k_wm = ozu.nans(diff_4k.shape)
diff_1k_wm[wm_idx] = diff_1k[wm_idx]
diff_2k_wm[wm_idx] = diff_2k[wm_idx]
diff_4k_wm[wm_idx] = diff_4k[wm_idx]


```
